# **Medidas Separatrizes: Conceito e Aplicação**  

As **medidas separatrizes** são estatísticas que dividem um conjunto ordenado de dados em partes iguais. Essas medidas ajudam a analisar a **distribuição dos dados**, facilitando a interpretação de valores e permitindo comparações entre diferentes grupos ou percentuais da amostra.  

Elas são amplamente utilizadas em estatística descritiva e aplicadas em áreas como economia, educação, análise de desempenho e ciência de dados.  


## **1. Tipos de Medidas Separatrizes**  

As principais medidas separatrizes são:  

1. **Quartis (Qi)** → Dividem os dados em **4 partes** de 25%.  
2. **Decis (Di)** → Dividem os dados em **10 partes** de 10%.  
3. **Percentis (Pi)** → Dividem os dados em **100 partes** de 1%.  

Cada uma dessas medidas determina um valor que separa os dados em frações específicas, permitindo análises como:  
- Comparar desempenhos individuais dentro de um grupo.  
- Identificar valores extremos (outliers).  
- Avaliar a distribuição de um fenômeno (salários, notas, tempos de resposta etc.).  



### **Fórmula Geral**  

Para qualquer medida separatriz, o valor $P_k$ (onde $k$ é o percentil, decil ou quartil desejado) pode ser obtido pela fórmula:

$P_k = \left( \frac{k}{m} \right) \times (n+1)$

onde:
-$k$ = número da medida separatriz desejada (exemplo: 1º quartil, 3º decil, 90º percentil).  
-$m$ = número total de divisões (4 para quartis, 10 para decis, 100 para percentis).  
-$n$ = total de elementos no conjunto de dados ordenado.  

Se o índice encontrado não for um número inteiro, deve-se **interpolar** entre os valores adjacentes.

### Exemplo de Cálculo de Medidas Separatrizes em Python

Vamos usar o conjunto de dados e calcular as medidas separatrizes.

````python
import numpy as np
import pandas as pd

# Conjunto de dados
dados = [10, 20, 35, 50, 60, 75, 80, 90, 100, 120, 130, 150]

# Convertendo para um array do NumPy
dados_array = np.array(dados)

# Calculando as Medidas Separatrizes
q1 = np.percentile(dados_array, 25)  # Primeiro quartil
q2 = np.percentile(dados_array, 50)  # Mediana (segundo quartil)
q3 = np.percentile(dados_array, 75)  # Terceiro quartil

# Calculando Decis
d1 = np.percentile(dados_array, 10)  # Decil 1
d5 = np.percentile(dados_array, 50)  # Decil 5 (Mediana)
d9 = np.percentile(dados_array, 90)  # Decil 9

# Calculando Percentis
p10 = np.percentile(dados_array, 10)  # Percentil 10
p50 = np.percentile(dados_array, 50)  # Percentil 50 (mediana)
p90 = np.percentile(dados_array, 90)  # Percentil 90

# Exibindo os resultados
print("Quartis:")
print(f"Q1: {q1}")
print(f"Q2 (Mediana): {q2}")
print(f"Q3: {q3}")

print("\nDecis:")
print(f"D1: {d1}")
print(f"D5: {d5}")
print(f"D9: {d9}")

print("\nPercentis:")
print(f"Percentil 10: {p10}")
print(f"Percentil 50: {p50}")
print(f"Percentil 90: {p90}")
````

# **Medidas de dispersão**

As medidas de dispersão são estatísticas descritivas que quantificam o grau de variação ou dispersão de um conjunto de dados em relação à sua tendência central. Enquanto medidas de tendência central, como média, mediana e moda, fornecem um valor representativo dos dados, as medidas de dispersão indicam o quanto os valores individuais diferem desse ponto central. A análise dessas medidas é essencial para compreender a variabilidade dos dados e para comparar diferentes distribuições estatísticas.


## **1. Amplitude**  
A **amplitude** é uma das medidas de dispersão mais simples em estatística, sendo utilizada para quantificar a diferença entre o maior e o menor valor de um conjunto de dados. Apesar de sua simplicidade, a amplitude desempenha um papel fundamental na análise exploratória dos dados, permitindo uma **avaliação inicial da variabilidade** dentro de uma distribuição.  

$Amplitude = Valor Máximo - Valor Mínimo$

---



### **1. Definição e Cálculo da Amplitude**  

A amplitude pode ser definida matematicamente da seguinte forma:  

$\text{Amplitude} = X_{\max} - X_{\min}$  

onde:  
-$X_{\max}$ representa o maior valor do conjunto de dados.  
-$X_{\min}$ representa o menor valor do conjunto de dados.  

A amplitude expressa a **distância total** coberta pelos valores dentro de uma distribuição. Se os dados forem muito dispersos, a amplitude será grande; se forem concentrados, a amplitude será pequena.  

````python
import numpy as np

dados = [5, 8, 12, 14, 17, 21, 24]

amplitude = np.ptp(dados)  # ptp retorna a diferença entre max e min

print(f"Amplitude: {amplitude}")
````

#### Como Interpretar a Amplitude?

A amplitude é uma medida simples e rápida para identificar a variabilidade dos dados. Quanto maior a amplitude, maior a dispersão entre os valores, e isso pode indicar que os dados são mais variados.

#### Limitação da Amplitude:
Embora a amplitude seja útil, ela é muito sensível a outliers (valores extremos). Se um valor for muito distante dos outros, a amplitude pode dar uma visão distorcida da distribuição dos dados. Para evitar isso, medidas como desvio padrão ou mediana podem ser mais adequadas para dados com outliers.

## **2. Desvio Médio**  

O **Desvio Médio** é uma medida de dispersão que indica **o afastamento médio dos valores de um conjunto de dados em relação à sua média aritmética**. Ele expressa, em termos absolutos, o quanto os dados se espalham em torno da média, sendo útil para entender a variabilidade dos dados sem considerar a direção das diferenças.  

### **1. Conceito e Definição**  

O desvio médio é calculado como a média das diferenças absolutas entre cada valor do conjunto de dados e a média aritmética dos valores. Sua fórmula geral é dada por:  

#### **1.1. Para uma População**  

$DM = \frac{\sum |X_i - \mu|}{N}$

onde:  
-$DM$ = Desvio médio  
-$X_i$ = Cada valor do conjunto de dados  
-$\mu$ = Média populacional  
-$N$ = Número total de elementos da população  

#### **1.2. Para uma Amostra**  

$DM = \frac{\sum |X_i - \bar{X}|}{n}$

onde:  
-$\bar{X}$ = Média amostral  
-$n$ = Número de elementos da amostra  

O uso da **média aritmética** como referência para medir as diferenças permite entender como os dados se comportam em relação ao valor central.  

#### Exemplo python

````python
import numpy as np

# Exemplo de dados
dados = [10, 12, 23, 23, 16, 23, 21, 16]

# Passo 1: Calcular a média
media = np.mean(dados)

# Passo 2: Calcular as diferenças absolutas em relação à média
diferencas_absolutas = np.abs(dados - media)

# Passo 3: Calcular o desvio médio
desvio_medio = np.mean(diferencas_absolutas)

# Exibir os resultados
print(f"Dados: {dados}")
print(f"Média: {media}")
print(f"Diferenças absolutas: {diferencas_absolutas}")
print(f"Desvio Médio: {desvio_medio}")
````

#### **Interpretação:**
A média dos dados é 18.5.
As diferenças absolutas entre cada valor e a média variam, e o desvio médio é 4.5. Isso significa que, em média, os valores dos dados estão a 4.5 unidades de distância da média.

## **3. Variância**  

A **variância** é uma medida estatística que expressa a dispersão dos dados em relação à média. Em outras palavras, ela quantifica **o quão distantes os valores estão da média** de um conjunto de dados. Essa medida é fundamental para entender a estabilidade e a confiabilidade de um conjunto de informações, sendo amplamente utilizada em estatística descritiva, inferência estatística, machine learning e diversas áreas aplicadas.  

### **1. Conceito de Variância**  

A variância mede a média dos **quadrados das diferenças** entre cada valor e a média aritmética do conjunto de dados. Ela é representada pelas seguintes fórmulas:  

- **Para uma população inteira** (variância populacional):  

$\sigma^2 = \frac{\sum (X_i - \mu)^2}{N}$

- **Para uma amostra** (variância amostral):  

$s^2 = \frac{\sum (X_i - \bar{X})^2}{n - 1}$

onde:  
-$X_i$ representa cada valor do conjunto de dados,  
-$\mu$ é a média populacional,  
-$\bar{X}$ é a média amostral,  
-$N$ é o número total de elementos da população,  
-$n$ é o número total de elementos da amostra,  
-$\sigma^2$ é a variância populacional,  
-$s^2$ é a variância amostral.  

A principal diferença entre **variância populacional** e **variância amostral** é o denominador. No caso amostral, subtrai-se **1** do número de elementos (graus de liberdade) para corrigir o viés da estimativa, garantindo que a variância amostral seja uma estimativa não tendenciosa da variância populacional.

#### Exemplo Prático de Cálculo da Variância em Python

Vamos calcular a variância de um conjunto de dados usando Python, primeiro manualmente e depois utilizando funções de bibliotecas como NumPy.


````python
import numpy as np

# Dados de exemplo
dados = [12, 15, 14, 10, 18, 16, 19]

# Passo 1: Calcular a média (mu ou x_bar)
media = np.mean(dados)

# Passo 2: Calcular a soma dos quadrados das diferenças em relação à média
soma_dos_quadrados = sum([(x - media)**2 for x in dados])

# Passo 3: Calcular a variância
variancia = soma_dos_quadrados / len(dados)  # Para variância populacional

print(f"Média: {media}")
print(f"Soma dos quadrados das diferenças: {soma_dos_quadrados}")
print(f"Variância: {variancia}")
````

### Cálculo Usando NumPy

O NumPy fornece funções prontas para calcular a variância de maneira mais eficiente. Vamos usar np.var() para calcular a variância diretamente:

````python
# Calculando a variância com NumPy
variancia_numpy = np.var(dados)

print(f"Variância (usando NumPy): {variancia_numpy}")
````

````python
# Calculando a variância amostral com NumPy
variancia_amostral_numpy = np.var(dados, ddof=1)

print(f"Variância Amostral (usando NumPy): {variancia_amostral_numpy}")
````

### Visualizando a Variância

Para ter uma noção visual da variância, podemos plotar um gráfico de dispersão (scatter plot) mostrando a média e as diferenças em relação a ela.

````python
import matplotlib.pyplot as plt

# Plotando os dados e a média
plt.scatter(dados, [0]*len(dados), color='blue', label="Dados")
plt.axvline(media, color='red', label=f'Média: {media}')

# Adicionando elementos no gráfico
plt.title("Distribuição dos Dados e Média")
plt.xlabel("Valores")
plt.legend()
plt.show()
````

## **4. Desvio Padrão**  

O **desvio padrão** é uma das medidas de dispersão mais importantes da estatística, pois indica o grau de variação ou dispersão dos dados em relação à média. Essa métrica é amplamente utilizada em diversas áreas, como ciência de dados, economia, engenharia, ciências sociais e análise de riscos, pois fornece uma visão quantitativa da estabilidade e previsibilidade de um conjunto de dados.  


### **1. Definição do Desvio Padrão**  

O **desvio padrão** (representado por$\sigma$ para populações e$s$ para amostras) mede **o quanto os valores de um conjunto de dados se afastam da média**. Ele é definido como a raiz quadrada da variância, permitindo que a dispersão seja expressa na mesma unidade dos dados originais.  

A fórmula do desvio padrão **populacional** ($\sigma$) é:  

$\sigma = \sqrt{\frac{\sum (X_i - \mu)^2}{N}}$  

Onde:  
-$X_i$ representa cada valor do conjunto de dados,  
-$\mu$ é a média populacional,  
-$N$ é o número total de elementos na população.  

Para amostras, o **desvio padrão amostral** ($s$) é calculado com um pequeno ajuste, substituindo$N$ por$n - 1$ (graus de liberdade):  

$s = \sqrt{\frac{\sum (X_i - \bar{X})^2}{n - 1}}$  

Esse ajuste é necessário para corrigir a subestimação da variabilidade da população quando usamos uma amostra.

#### Exemplo 1

````python
import numpy as np

# Dados de exemplo
dados = [10, 12, 23, 23, 16, 23, 21, 17, 18, 20]

# Calculando a média
media = np.mean(dados)

# Calculando o desvio padrão
desvio_padrao = np.std(dados)

# Exibindo os resultados
print(f'Média dos dados: {media}')
print(f'Desvio padrão dos dados: {desvio_padrao}')
````

##### Interpretação:
**Média**: O valor médio dos dados é 19.3, o que significa que, em média, os dados estão em torno de 19.3.  
**Desvio padrão**: O valor de 4.78 indica que os dados variam cerca de 4.78 unidades para cima ou para baixo da média.

#### **Exemplo 2: Comparação de Dois Conjuntos de Dados**

````python
# Conjunto de dados 1 (mais dispersos)
dados_1 = [5, 10, 15, 20, 25]

# Conjunto de dados 2 (menos dispersos)
dados_2 = [10, 11, 12, 13, 14]

# Calculando os desvios padrões
desvio_padrao_1 = np.std(dados_1)
desvio_padrao_2 = np.std(dados_2)

# Exibindo os resultados
print(f'Desvio padrão do conjunto 1: {desvio_padrao_1}')
print(f'Desvio padrão do conjunto 2: {desvio_padrao_2}')
````

#### **Exemplo 3: Cálculo Manual do Desvio Padrão**

````python
# Dados de exemplo
dados = [10, 12, 23, 23, 16, 23, 21, 17, 18, 20]

# Calculando a média
media = sum(dados) / len(dados)

# Calculando a soma dos quadrados das diferenças da média
soma_dos_quadrados = sum([(x - media)**2 for x in dados])

# Calculando o desvio padrão
desvio_padrao_manual = (soma_dos_quadrados / len(dados))**0.5

# Exibindo os resultados
print(f'Média dos dados: {media}')
print(f'Desvio padrão (manual): {desvio_padrao_manual}')
````

# Identificacao de outliers

### O que são Outliers?

**Outliers** são valores que **fogem muito do padrão** do restante dos dados.  
Eles estão **muito acima ou muito abaixo da média** e podem **indicar erros de medição, casos especiais, ou situações reais incomuns.**

> Em outras palavras: **outliers são pontos "estranhos" que não seguem o comportamento geral do conjunto.**

---


### **Regra do Intervalo Interquartil (IQR)**

Esse método utiliza os **quartis** (Q1 e Q3) para identificar os limites de valores aceitáveis.

#### Fórmulas:
- IQR = Q3 - Q1
- Limite inferior = Q1 - 1.5 × IQR  
- Limite superior = Q3 + 1.5 × IQR

> **Valores fora desse intervalo são considerados outliers.**

#### Exemplo:
```
Dados: [10, 12, 13, 13, 14, 15, 16, 17, 30]
Q1 = 13 | Q3 = 16 | IQR = 3
Limites: [13 - 1.5×3 = 8.5 ; 16 + 1.5×3 = 20.5]
➡️ O valor 30 é um outlier.
```

---



````python
import numpy as np
import matplotlib.pyplot as plt

# Exemplo de dados (com alguns outliers)
dados = np.array([1, 7, 7, 8, 9, 10, 10, 11, 11, 12, 12, 13, 13, 14, 15, 16, 17, 30])

# Calcular os quartis
q1 = np.percentile(dados, 25)
q3 = np.percentile(dados, 75)
iqr = q3 - q1

# Limites para detecção de outliers
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

# Identificar outliers
outliers = dados[(dados < limite_inferior) | (dados > limite_superior)]

# Valores dos "bigodes" (sem outliers)
dados_sem_outliers = dados[(dados >= limite_inferior) & (dados <= limite_superior)]
bigode_inferior = np.min(dados_sem_outliers)
bigode_superior = np.max(dados_sem_outliers)

# Exibir resultados
print(f"Q1: {q1}")
print(f"Q3: {q3}")
print(f"IQR: {iqr}")
print(f"Limite Inferior: {limite_inferior}")
print(f"Limite Superior: {limite_superior}")
print(f"Bigode Inferior: {bigode_inferior}")
print(f"Bigode Superior: {bigode_superior}")
print(f"Outliers encontrados: {outliers}")

# Gráfico de boxplot com largura aumentada
plt.figure(figsize=(10, 4))
plt.boxplot(dados, vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"), widths=0.5)
plt.title("Boxplot com IQR, Bigodes e Outliers")
plt.xlabel("Valores")
plt.grid(True)

# Adicionar valores de Q1, Q3, Mediana
mediana = np.median(dados)
plt.text(q1, 1.05, f'Q1: {q1:.2f}', ha='center', color='blue', fontsize=9)
plt.text(mediana, 1.10, f'Mediana: {mediana:.2f}', ha='center', color='green', fontsize=9)
plt.text(q3, 1.05, f'Q3: {q3:.2f}', ha='center', color='blue', fontsize=9)

# Anotar os bigodes
plt.text(bigode_inferior, 0.90, f'{bigode_inferior}', ha='center', color='purple', fontsize=9)
plt.text(bigode_superior, 0.90, f'{bigode_superior}', ha='center', color='purple', fontsize=9)

# Anotar os outliers
for outlier in outliers:
    plt.text(outlier, 1.02, f'{outlier}', ha='center', color='red', fontsize=9)

# Salvar e exibir
plt.savefig("exemplo_iqr_bigodes.png", dpi=300)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Exemplo de dados (com alguns outliers)
dados = np.array([2,5,7,8,10,12,15,18,25,30])

# Calcular os quartis
q1 = np.percentile(dados, 25, method='midpoint')
q3 = np.percentile(dados, 75, method='midpoint')
iqr = q3 - q1

# Limites para detecção de outliers
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

# Identificar outliers
outliers = dados[(dados < limite_inferior) | (dados > limite_superior)]

# Valores dos "bigodes" (sem outliers)
dados_sem_outliers = dados[(dados >= limite_inferior) & (dados <= limite_superior)]
bigode_inferior = np.min(dados_sem_outliers)
bigode_superior = np.max(dados_sem_outliers)

# Exibir resultados
print(f"Q1: {q1}")
print(f"Q3: {q3}")
print(f"IQR: {iqr}")
print(f"Limite Inferior: {limite_inferior}")
print(f"Limite Superior: {limite_superior}")
print(f"Bigode Inferior: {bigode_inferior}")
print(f"Bigode Superior: {bigode_superior}")
print(f"Outliers encontrados: {outliers}")

# Gráfico de boxplot com largura aumentada
plt.figure(figsize=(10, 4))
plt.boxplot(dados, vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"), widths=0.5)
plt.title("Boxplot com IQR, Bigodes e Outliers")
plt.xlabel("Valores")
plt.grid(True)

# Adicionar valores de Q1, Q3, Mediana
mediana = np.median(dados)
plt.text(q1, 1.05, f'Q1: {q1:.2f}', ha='center', color='blue', fontsize=9)
plt.text(mediana, 1.10, f'Mediana: {mediana:.2f}', ha='center', color='green', fontsize=9)
plt.text(q3, 1.05, f'Q3: {q3:.2f}', ha='center', color='blue', fontsize=9)

# Anotar os bigodes
plt.text(bigode_inferior, 0.90, f'{bigode_inferior}', ha='center', color='purple', fontsize=9)
plt.text(bigode_superior, 0.90, f'{bigode_superior}', ha='center', color='purple', fontsize=9)

# Anotar os outliers
for outlier in outliers:
    plt.text(outlier, 1.02, f'{outlier}', ha='center', color='red', fontsize=9)

# Salvar e exibir
plt.savefig("exemplo_iqr_bigodes_sem_outliers.png", dpi=300)
plt.show()
````

## Identificação de Outliers com a **Regra Empírica (Regra dos Três Sigmas)**

### O que é a Regra Empírica?

A **Regra Empírica** se aplica a **distribuições normais** (ou aproximadamente normais), ou seja, aquelas com formato de sino (a famosa curva "gaussiana").

Essa regra descreve como os dados se distribuem **em torno da média**, considerando o **desvio padrão** como unidade de medida.

---



### A Regra 68-95-99.7

De acordo com essa regra:

- ✅ Cerca de **68%** dos dados caem dentro de **±1 desvio padrão da média**.
- ✅ Cerca de **95%** dos dados caem dentro de **±2 desvios padrão da média**.
- ✅ Cerca de **99,7%** dos dados caem dentro de **±3 desvios padrão da média**.

---

### Como isso ajuda a identificar outliers?

Se um valor estiver:

- **Fora do intervalo de ±3σ da média**, ele é considerado **um outlier extremo**.
- **Entre ±2σ e ±3σ**, pode ser um **outlier moderado**, dependendo do contexto.

---

### Fórmulas práticas

Seja:

- μ = média
- σ = desvio padrão

Você pode calcular os limites assim:

- **Limite inferior**: μ - 3σ  
- **Limite superior**: μ + 3σ

> Qualquer valor fora desse intervalo é considerado **um outlier com base na Regra Empírica**.

---

### Quando usar essa regra?

- Quando seus dados seguem (ou se aproximam de) uma **distribuição normal**.
- Quando você quer uma **forma rápida e prática de identificar outliers**.
- Quando deseja **complementar** métodos como **Z-score** ou **boxplot** com uma visão mais teórica.

---

### Dica prática:

Você pode usar o **Z-score** para calcular diretamente isso:

```python
from scipy.stats import zscore
import numpy as np

dados = np.array([100, 105, 110, 115, 120, 130, 140, 160])
z_scores = zscore(dados)

outliers = dados[np.abs(z_scores) > 3]
print("Outliers:", outliers)
```

---

### Conclusão

A **Regra Empírica** é uma forma poderosa e simples de **identificar outliers**, especialmente em distribuições normais.  
Ela mostra que **valores muito distantes da média são raros** — e por isso devem ser observados com atenção!


````python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Configurações da distribuição normal
mu = 100  # Média
sigma = 15  # Desvio padrão

# Gerar dados da distribuição normal
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = stats.norm.pdf(x, mu, sigma)

# Mostrar um resumo dos dados
print("Amostra dos dados gerados para o gráfico:")
for xi, yi in zip(x[::100], y[::100]):  # Mostra a cada 100 pontos
    print(f"x = {xi:.2f}, y = {yi:.5f}")

# Criar o gráfico
plt.figure(figsize=(10, 5))
plt.plot(x, y, label="Distribuição Normal", color="black")

# Destacar regiões da Regra Empírica
for i, alpha in zip(range(1, 4), [0.3, 0.2, 0.1]):
    plt.fill_between(x, y, where=(mu - i*sigma <= x) & (x <= mu + i*sigma),
                     color="blue", alpha=alpha, label=f"{68 if i == 1 else 95 if i == 2 else 99.7}% dentro de {i}σ")

# Adicionar linhas verticais para a média e desvios padrão
plt.axvline(mu, color='red', linestyle='dashed', label='Média')
plt.axvline(mu - sigma, color='gray', linestyle='dashed', label='1 Desvio Padrão')
plt.axvline(mu + sigma, color='gray', linestyle='dashed')
plt.axvline(mu - 2*sigma, color='gray', linestyle='dashed', label='2 Desvios Padrão')
plt.axvline(mu + 2*sigma, color='gray', linestyle='dashed')
plt.axvline(mu - 3*sigma, color='gray', linestyle='dashed', label='3 Desvios Padrão')
plt.axvline(mu + 3*sigma, color='gray', linestyle='dashed')


# Salvar dados da distribuição em CSV e Excel
dados = pd.DataFrame({'Valor': x, 'Densidade': y})

# CSV
dados.to_csv('dados_distribuicao.csv', index=False)

# Excel
dados.to_excel('dados_distribuicao.xlsx', index=False)

# Configurações do gráfico
plt.title("Regra Empírica (68-95-99.7) na Distribuição Normal")
plt.xlabel("Valores")
plt.ylabel("Densidade de Probabilidade")
plt.legend()
plt.show()
````

# Exemplo de simetria


---

### 📘 1. O que é uma **distribuição simétrica**?

É quando os dados estão **distribuídos igualmente** ao redor da média. Isso significa que:

- **Média ≈ Mediana ≈ Moda**
- O gráfico tem **forma de sino** (curva normal).
- Os dois lados do gráfico são **espelhados**.

### ✅ Exemplo prático (simulado):

Alturas de uma população:
```python
dados = [165, 167, 168, 170, 172, 173, 175]
```

- Média: 170
- Mediana: 170
- Moda: não há moda clara, os dados são bem centrados

📈 Gráfico: seria como uma **curva de sino bem centralizada**.  
Imagine esse gráfico como um morro com lados iguais:

```
          *
        *   *
      *       *
    *           *
  *               *
```

---

## 📕 2. O que é uma **distribuição assimétrica**?

É quando os dados estão **mais concentrados de um lado**, criando uma **cauda longa do outro lado**.

Existem dois tipos:

### 🔸 Assimetria **à direita** (positiva):
- **Média > Mediana > Moda**
- Muitos valores pequenos, poucos valores grandes

#### Exemplo:
Renda mensal (em reais):
```python
dados = [1000, 1100, 1200, 1300, 1500, 4000, 9000]
```

- Média: alta, puxada por 9000
- Mediana: menor que a média
- Moda: valores mais baixos

📈 Gráfico: a cauda vai para a **direita**:

```
      *
     * *
    *   *
   *     *
  *       *
 *          *
*             *
```

---

### 🔹 Assimetria **à esquerda** (negativa):
- **Média < Mediana < Moda**
- Muitos valores altos, poucos muito baixos

#### Exemplo:
Notas de uma turma onde quase todo mundo foi bem:
```python
dados = [2, 8, 8, 9, 9, 9, 10]
```

- Média: puxada para baixo por 2
- Mediana: maior que a média
- Moda: 9

📈 Gráfico: a cauda vai para a **esquerda**:

```
            *
          *   *
        *       *
      *           *
    *               *
  *                   *
*                      *
```

---

## Dica para identificar:

| Tipo              | Média vs Mediana | Cauda puxa para... |
|-------------------|------------------|--------------------|
| Simétrica         | Média = Mediana  | Nenhum lado        |
| Assimétrica direita | Média > Mediana  | Direita (valores grandes) |
| Assimétrica esquerda | Média < Mediana | Esquerda (valores pequenos) |

---

Se quiser, eu posso fazer gráficos de verdade em Python com seus dados ou dados fictícios. Quer que eu mostre visualmente isso?

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Estilo visual
sns.set(style="whitegrid")

# Gerar dados
np.random.seed(42)
simetrica = np.random.normal(loc=50, scale=10, size=1000)
assimetrica_direita = np.random.exponential(scale=10, size=1000)
assimetrica_esquerda = -np.random.exponential(scale=10, size=1000) + 60

# Função auxiliar para plotar as linhas
def plot_medidas(ax, data):
    media = np.mean(data)
    mediana = np.median(data)
    moda = stats.mode(data, keepdims=True).mode[0]

    ax.axvline(media, color='red', linestyle='--', label='Média')
    ax.axvline(mediana, color='green', linestyle='-', label='Mediana')
    ax.axvline(moda, color='blue', linestyle='-.', label='Moda')

    ax.legend()

# Criar os gráficos
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Simétrica
sns.histplot(simetrica, kde=True, bins=30, ax=axes[0], color='skyblue')
axes[0].set_title("Distribuição Simétrica")
axes[0].set_xlabel("Valores")
axes[0].set_ylabel("Frequência")
plot_medidas(axes[0], simetrica)

# Assimétrica à direita
sns.histplot(assimetrica_direita, kde=True, bins=30, ax=axes[1], color='orange')
axes[1].set_title("Assimétrica à Direita")
axes[1].set_xlabel("Valores")
plot_medidas(axes[1], assimetrica_direita)

# Assimétrica à esquerda
sns.histplot(assimetrica_esquerda, kde=True, bins=30, ax=axes[2], color='lightgreen')
axes[2].set_title("Assimétrica à Esquerda")
axes[2].set_xlabel("Valores")
plot_medidas(axes[2], assimetrica_esquerda)

# >>> SALVA A IMAGEM COMO PNG <<<
plt.savefig("dados_simentricos_assimetricos.png", dpi=300)

plt.tight_layout()
plt.show()
````

## Z-score



````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Conjunto de dados (notas de alunos)
notas = np.array([60, 70, 75, 80, 85, 90, 100, 30])  # Incluímos um outlier (30)

# Passo 1: Calcular média e desvio padrão
media = np.mean(notas)
desvio_padrao = np.std(notas)

# Passo 2: Calcular Z-score para cada valor
z_scores = (notas - media) / desvio_padrao

# Passo 3: Identificar outliers (Z-score maior que 3 ou menor que -3)
outliers = notas[np.abs(z_scores) > 3]

# Exibir resultados
print(f"Média: {media:.2f}")
print(f"Desvio Padrão: {desvio_padrao:.2f}")
print("Z-scores:", z_scores)
print("Outliers identificados:", outliers)

# ---- VISUALIZAÇÃO ----
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Gráfico de dispersão com destaque para outliers
axs[0].scatter(range(len(notas)), notas, label="Dados", color="blue")
axs[0].scatter(np.where(np.abs(z_scores) > 3), outliers, label="Outliers", color="red", marker="o", s=100)
axs[0].axhline(media, color="green", linestyle="--", label="Média")
axs[0].set_title("Dispersão das Notas")
axs[0].set_xlabel("Aluno")
axs[0].set_ylabel("Nota")
axs[0].legend()

# Boxplot para visualizar outliers
sns.boxplot(y=notas, ax=axs[1])
axs[1].set_title("Boxplot das Notas")

plt.tight_layout()
plt.show()
````

## Data analysis

### Subtask:
Calculate measures of central tendency and dispersion for the numerical columns in `df_analysis`.


**Reasoning**:
Calculate the measures of central tendency, dispersion, and quartiles for the numerical columns in `df_analysis` and store them in a new DataFrame called `summary_stats`.



````python
import pandas as pd
import numpy as np

# Calculate measures of central tendency
mean_values = df_analysis.mean()
median_values = df_analysis.median()
mode_values = df_analysis.mode().iloc[0]  # Get the first mode if multiple exist

# Calculate measures of dispersion
range_values = df_analysis.max() - df_analysis.min()
variance_values = df_analysis.var()
std_values = df_analysis.std()

# Calculate quartiles and IQR
q1_values = df_analysis.quantile(0.25)
q2_values = df_analysis.quantile(0.50)  # Median
q3_values = df_analysis.quantile(0.75)
iqr_values = q3_values - q1_values

# Create the summary statistics DataFrame
summary_stats = pd.DataFrame({
    'mean': mean_values,
    'median': median_values,
    'mode': mode_values,
    'range': range_values,
    'variance': variance_values,
    'std': std_values,
    'Q1': q1_values,
    'Q2': q2_values,
    'Q3': q3_values,
    'IQR': iqr_values
})

# Set the column names of df_analysis as index for summary_stats
summary_stats.index = df_analysis.columns

display(summary_stats)
````
